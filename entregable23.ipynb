```python
import pandas as pd
import seaborn as snss
```

```python
import numpy
from matplotlib import pyplot
from scipy.stats import linregress
from scipy import stats
import matplotlib.pyplot as plt
from pylab import *

```

```python
print("Daniel Dataset LATERAL")
df1 = pd.read_csv("datasets/Daniel_IMU_X_LATERAL.txt", skiprows=4, delimiter="\t")
df2 = pd.read_csv("datasets/Daniel_MOCAP_X_LATERAL.txt", skiprows=4, delimiter="\t")
data_imu = abs(df1.iloc[:,1:].values)
data_imu = data_imu.reshape(-1)
data_imu.shape
data_mocap = abs(df2.iloc[:,1:].values)
data_mocap = data_mocap.reshape(-1)
data_mocap.shape
coef = np.polyfit(data_imu, data_mocap,1)
print(coef)
poly1d_fn = np.poly1d(coef) 
plt.xlabel("IMU Lateral (X)")
plt.ylabel("MOCAP Lateral (X)")
plt.plot(data_imu, data_mocap, 'yo', data_imu, poly1d_fn(data_imu), '--k')
slope, intercept, r_value, p_value, std_err = stats.linregress(data_imu, data_mocap)

print('Slope: ',slope,'\nIntercept: ',intercept)
print("R2: ", r_value**2)
print(f"p_value: {p_value} std_err {std_err}")
print("IMU_Mean: ", data_imu.mean(), " IMU_SD: ",data_imu.std())
print("Mocap_Mean: ", data_mocap.mean(), " MOCAP_SD: ", data_mocap.std())
```
